```python
import random

class ParkingLot:
    def __init__(self, capacity):
        self.capacity = capacity
        self.available_spaces = capacity

    def park_vehicle(self):
        if self.available_spaces > 0:
            self.available_spaces -= 1
            return True
        else:
            return False

    def leave_parking(self):
        if self.available_spaces < self.capacity:
            self.available_spaces += 1
            return True
        else:
            return False

def simulate_smart_parking():
    capacity = 50
    parking_lot = ParkingLot(capacity)

    for _ in range(100):
        action = random.choice(["park", "leave"])
        if action == "park":
            if parking_lot.park_vehicle():
                print("Vehicle parked.")
            else:
                print("Parking full.")
        else:
            if parking_lot.leave_parking():
                print("Vehicle left.")
            else:
                print("No vehicles to leave.")

simulate_smart_parking()
```
